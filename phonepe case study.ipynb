#2.6 Analyze app usage trends

```python
# calculate the total bumber of app opens over the years and quaters for each state. Display the result.
open_app= (pd_state_t_u.groupby(['State','Year','Quarter'])['App Opens'].sum().reset_index().sort_values(by=['State','Year','Quarter']))
open_app
```

```python
#2identify trends in app usage by creating a line plot showing th enumber of app open over the time for a selected state.
# Select the state to plot
State = 'Karnataka'  # Change this to the desired state

# Filter for the selected state
df = pd_state_t_u[pd_state_t_u['State'] == State].copy()

# Convert quarter number (e.g., 1.0) to string like 'Q1'
df['Quarter'] = 'Q' + df['Quarter'].astype(int).astype(str)

# Combine year and quarter into a time label
df['Time'] = df['Year'].astype(int).astype(str) + ' ' + df['Quarter']

# Sort for proper plotting
df = df.sort_values(by=['Year', 'Quarter'])

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df['Time'], df['App Opens'], marker='o')
plt.title(f'App Opens Over Time in {State}')
plt.xlabel('Time')
plt.ylabel('Number of App Opens')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

# 2.7 Distribution of transaction type

```python
#1 create a bar chart showing the distribution of different transaction types for each state for the most recent quarter in the data set
## find latest year
latest_year=pd_txn['Year'].max()
# latest quarter
latest_quarter=pd_txn[pd_txn['Year']==latest_year]['Quarter'].max()
# filter most latest quarter
most_r_qu=pd_txn[(pd_txn['Year']==latest_year) & (pd_txn['Quarter']==latest_quarter)]
#group by state , tranasaction type , summing transaction
grouped=most_r_qu.groupby(['State','Transaction Type'])['Transactions'].sum().reset_index()

#plot
plt.figure(figsize=(13,7))
sns.barplot(data=grouped,x='State',y='Transactions',hue='Transaction Type')
plt.title(f"Transaction Distribution by type for each state-{latest_year}Q{latest_quarter}")
plt.xlabel("State")
plt.ylabel("Number of Transaction")
plt.xticks(rotation=45)
plt.legend(title='Transaction Type')
plt.tight_layout()
plt.show()
```
