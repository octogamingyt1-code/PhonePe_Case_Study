# 4.3 Average transaction amount per user

#  1 . merge relevant datasets to calculate the average transaction amount per user for each state. Display the results in a tabular format..

```python
import pandas as pd

# Step 1: Clean column names
pd_state_t_u.columns = pd_state_t_u.columns.str.strip()
pd_sate_dev.columns = pd_sate_dev.columns.str.strip()

# Step 2: Aggregate total registered users per State-Year-Quarter from pd_sate_dev
user_summary = pd_sate_dev.groupby(['State', 'Year', 'Quarter'])['Registered Users'].sum().reset_index()
user_summary.rename(columns={'Registered Users': 'Total Registered Users'}, inplace=True)

# Step 3: Merge with transaction data on State-Year-Quarter
merged = pd.merge(pd_state_t_u, user_summary, on=['State', 'Year', 'Quarter'], how='inner')

# Step 4: Compute average transaction amount per user
merged['avg_transaction_per_user'] = merged['Amount (INR)'] / merged['Total Registered Users']

# Step 5: Group by state to get overall average per state
state_avg = merged.groupby('State')['avg_transaction_per_user'].mean().reset_index()
state_avg['avg_transaction_per_user'] = state_avg['avg_transaction_per_user'].round(2)

# Step 6: Display result
print(state_avg)
```

#  2 identify the top 5 states with the highest average transaction amount pe user and the top 5 state with the lowest average transaction amount oer user.

```python
top_5_state = state_avg.sort_values(by='avg_transaction_per_user' , ascending = False).head(5)

lowest_5_state= state_avg.sort_values(by = 'avg_transaction_per_user', ascending = True).head(5)

# display the result
print("Top 5 States by Average Transaction per User : ")
print(top_5_state)

print("\n Lowest 5 state by Average Transaction per User : ")
print(lowest_5_state)
```

#  6.2 Correlate demographic data with transaction data

### 1 Find correlation between demographic data (eg: Population Density) and Transaction Data(eg: Transaction volume). Summarize your findings.

```python
# standardize district and state names for proper join
pd_dis_demo['District']=pd_dis_demo['District'].str.strip().str.lower()
pd_dis_demo['State'] = pd_dis_demo['State'].str.strip().str.lower()

pd_dis_txn['District'] = pd_dis_txn['District'].str.strip().str.lower()
pd_dis_txn['State'] = pd_dis_txn['State'].str.strip().str.lower()

# merge two datasets
merge_df=pd.merge(pd_dis_demo,pd_dis_txn, on=['State','District'] , how='inner')

# If Density is missing, calculate it
if 'Density' not in merge_df.columns or merge_df['Density'].isna().any():
    merge_df['Density'] = merge_df['Population'] / merge_df['Area (sq km)']

# Select only numeric columns of interest
corr_df = merge_df[['Population', 'Density', 'Transactions', 'Amount (INR)']]

# Compute correlation matrix
correlation = corr_df.corr()
print(correlation)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation between Demographics and Transaction Metrics")
plt.tight_layout()
plt.show()
```
